```python
constant_feats = np.where(X.nunique() == 1)[0]

for col in X.columns[constant_feats]:
    print('Column {} has {} unique values : {}'.format(col, X[col].nunique(), X[col].unique()))
```

### ĞŸĞ¾ÑĞ¼Ğ¾Ñ‚Ñ€Ğ¸Ğ¼ Ñ Ğ½ĞµĞ¼Ğ½Ğ¾Ğ³Ğ¾ Ğ´Ñ€ÑƒĞ³Ğ¾Ğ¹ ÑÑ‚Ğ¾Ñ€Ğ¾Ğ½Ñ‹ Ğ¸ Ğ¾Ñ†ĞµĞ½Ğ¸Ğ¼ Ğ¼Ğ½Ğ¾Ğ¶ĞµÑÑ‚Ğ²Ğ° Ğ¾Ñ‚Ğ±Ñ€Ğ¾ÑˆĞµĞ½Ğ½Ñ‹Ñ… Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¾Ğ².

```python
# drop zeros 
zeros_cols = X.max()[X.max() == 0].index
negative_cols = []

for zero_col in zeros_cols:
    if df[zero_col].unique().any() != 0:
        display(df[zero_col].value_counts())
        negative_cols.append(zero_col)
```

```python
# delete from zeros negative predictors
zeros_cols = zeros_cols.drop(negative_cols)

# constant features is full of zeros_cols value
set(constant_feats) and set(zeros_cols) == set(zeros_cols)
```

### Ğ”Ğ°Ğ»ÑŒÑˆĞµ Ğ±ÑƒĞ´ĞµĞ¼ Ğ¸ÑĞ¿Ğ¾Ğ»ÑŒĞ·Ğ¾Ğ²Ğ°Ñ‚ÑŒ Ñ‚Ğ¾Ğ»ÑŒĞºĞ¾ Ğ¾Ğ´Ğ¸Ğ½ Ğ¸Ğ· ÑÑ‚Ğ¸Ñ… Ğ½Ğ°Ğ±Ğ¾Ñ€Ğ¾Ğ².
##### Ğ’Ğ¾Ğ·Ğ¼Ğ¾Ğ¶Ğ½Ğ¾, Ñ‡Ñ‚Ğ¾ Ğ² Ğ´Ğ°Ñ‚Ğ°ÑĞµÑ‚Ğµ ĞµÑÑ‚ÑŒ Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¸, Ñ€Ğ°Ğ²Ğ½Ñ‹Ñ… Ğ¿Ğ¾ ÑÑƒĞ¼Ğ¼Ğ°Ğ¼ Ğ¼Ğ¾Ğ´ÑƒĞ»ĞµĞ¹ Ğ·Ğ½Ğ°Ñ‡ĞµĞ½Ğ¸Ğ¹, Ğ¿Ğ¾ÑÑ‚Ğ¾Ğ¼Ñƒ Ğ¿Ğ¾ÑĞ¼Ğ¾Ñ‚Ñ€Ğ¸Ğ¼ Ğ³Ğ»Ğ°Ğ·Ğ°Ğ¼Ğ¸ Ğ½Ğ° ÑÑ‚Ğ¾.

```python
X_nonconst = X.drop(columns=X.columns[constant_feats])

X_nonconst.abs().sum()
```

```python
iter_abs = X_nonconst.abs().sum()
duplicates = pd.Index([])

for x in X_nonconst.abs().sum().iteritems():
    ind, value = x
    iter_abs.drop(ind, inplace=True)
    if (iter_abs == value).any():
        print(f'Equals: {ind} with {iter_abs[iter_abs == value].index.tolist()}, value {value}')
        duplicates = duplicates.append(other=iter_abs[iter_abs == value].index).drop_duplicates()
```

```python
lasso_clear = Lasso()
lasso_clear.fit(X_train[without_zeros_columns], y_train)

roc_score_clear = roc_auc_score(y_test, lasso_clear.predict(X_test[without_zeros_columns]))
print(f'{br}ROC AUC Score: {roc_score_clear}')
```

```python
from hyperopt import hp
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
import xgboost as xgb

dtrain = xgb.DMatrix(X.values, y.values)

params = {'colsample_bytree': 0.7, 
          'eta': 0.525, 
          'gamma': 0.65, 
          'max_depth': 10, 
          'min_child_weight': 1.0, 
          'n_estimators': 20, 
          'objective': 'binary:logistic', 
          'eval_metric': 'auc',
          'silent': 1,
          'subsample': 0.7}

xgb.cv(params, dtrain, num_boost_round=100, early_stopping_rounds=5)
```

```python
def score(params):
    print("Training with params : ")
    print(params)
    num_round = int(params['n_estimators'])
    del params['n_estimators']
    dtrain = xgb.DMatrix(X_train, label=y_train)
    dvalid = xgb.DMatrix(X_test, label=y_test)
    model = xgb.train(params, dtrain, num_round)
    predictions = model.predict(dvalid)
    score = roc_auc_score(y_test, predictions)
    br = '-'*124
    print(f'{br}\n\tScore of AUC: {score}\n{br}')
    return {'loss': score, 'status': STATUS_OK}

def optimize(trials):
    space = {
             'n_estimators' : hp.quniform('n_estimators', 10, 1000, 10),
             'eta' : hp.quniform('eta', 0.1, 0.2, 0.025),
             'max_depth' : hp.choice('max_depth', np.arange(3, 14, dtype=int)),
             'min_child_weight' : hp.quniform('min_child_weight', 1, 6, 1),
             'subsample' : hp.quniform('subsample', 0.7, 1, 0.05),
             'gamma' : hp.quniform('gamma', 0.6, 1, 0.05),
             'colsample_bytree' : hp.quniform('colsample_bytree', 0.7, 1, 0.05),
             'eval_metric': 'auc',
             'objective': 'binary:logistic',
             'nthread': 4,
             'silent' : 1
             }

    best = fmin(score, space, algo=tpe.suggest, trials=trials, max_evals=250)

    print(best)
    return best

    
trials = Trials()
best_opts = optimize(trials)
```

```python
best_params = {'colsample_bytree': 0.9500000000000001, 'eta': 0.125, 'gamma': 0.9, 'max_depth': 2,
                'min_child_weight': 1.0, 'n_estimators': 10.0, 'subsample': 0.75,              'eval_metric': 'auc',
                'objective': 'binary:logistic',
                'nthread': 4,
                'silent' : 1}

xgb.cv(best_params, dtrain, num_boost_round=100, early_stopping_rounds=15)                            
             
```
