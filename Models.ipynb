```python
X = df.drop(columns='y')
y = df.y
```

```python
# зато есть множество нулевых признаков

X[X.columns[X.sum() == 0]]
```

### Посмотрим на них несколько внимательнее

```python
constant_feats = np.where(X.nunique() == 1)[0]

for col in X.columns[constant_feats]:
    print('Column {} has {} unique values : {}'.format(col, X[col].nunique(), X[col].unique()))
```

### Посмотрим с немного другой стороны и оценим множества отброшенных признаков.

```python
# drop zeros 
zeros_cols = X.max()[X.max() == 0].index
negative_cols = []

for zero_col in zeros_cols:
    if df[zero_col].unique().any() != 0:
        display(df[zero_col].value_counts())
        negative_cols.append(zero_col)
```

```python
# delete from zeros negative predictors
zeros_cols = zeros_cols.drop(negative_cols)

# constant features is full of zeros_cols value
set(constant_feats) and set(zeros_cols) == set(zeros_cols)
```

### Дальше будем использовать только один из этих наборов.
##### Возможно, что в датасете есть признаки, равных по суммам модулей значений, поэтому посмотрим глазами на это.

```python
X_nonconst = X.drop(columns=X.columns[constant_feats])

X_nonconst.abs().sum()
```

```python
iter_abs = X_nonconst.abs().sum()
duplicates = pd.Index([])

for x in X_nonconst.abs().sum().iteritems():
    ind, value = x
    iter_abs.drop(ind, inplace=True)
    if (iter_abs == value).any():
        print(f'Equals: {ind} with {iter_abs[iter_abs == value].index.tolist()}, value {value}')
        duplicates = duplicates.append(other=iter_abs[iter_abs == value].index).drop_duplicates()
```

### Повторяющиеся по модулю значения есть. Например:
* feature-1494        34.000000 
* feature-1495        34.000000

```python
X_nonconst[duplicates].describe()
```

```python
# эти признаки выглядят довольно странно.

X_nonconst[X_nonconst.columns[X_nonconst.abs().sum().duplicated()]]
```

```python
# явный дубликат. 
X[[f'feature-{x}' for x in range(1494, 1496)]].plot(kind='scatter', x='feature-1494', y='feature-1495')
```

```python
without_zeros_columns = X.columns.drop(zeros_cols)
without_constants_columns = X_nonconst.columns
without_duplicates_columns = X_nonconst.drop(columns=duplicates).columns
```

```python
print(f'DataFrame columns count: {X.columns.shape[0]},\n\
Without zero-filled columns: {without_zeros_columns.shape[0]},\n\
Without constant-filled columns: {without_constants_columns.shape[0]},\n\
Without duplicated columns: {without_duplicates_columns.shape[0]}')
```

### Посмотрим, как ведет себя Лассо-регрессия

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    random_state=42)
```

```python
from sklearn.metrics import roc_auc_score

lr = Lasso()
lr.fit(X_train, y_train)

roc_score = roc_auc_score(y_test, lr.predict(X_test))
br = '-' * 64 + '\n'
print(f'{br}ROC AUC Score: {roc_score}')
```

```python
lasso_clear = Lasso()
lasso_clear.fit(X_train[without_zeros_columns], y_train)

roc_score_clear = roc_auc_score(y_test, lasso_clear.predict(X_test[without_zeros_columns]))
print(f'{br}ROC AUC Score: {roc_score_clear}')
```

### Лассо регрессия самостоятельно отбирает нулевые признаки как неинформативные, следовательно, первое предположение было верно.

```python
lasso_nonconst = Lasso()
lasso_nonconst.fit(X_train[without_constants_columns], y_train)

roc_score_nonconst = roc_auc_score(y_test, \
                                   lasso_nonconst.predict(X_test[without_constants_columns]))
print(f'{br}ROC AUC Score: {roc_score_nonconst}')
```

```python
lasso_d = Lasso()
lasso_d.fit(X_train[without_duplicates_columns], y_train)

roc_score_noduplicates = roc_auc_score(y_test, \
                                      lasso_d.predict(X_test[without_duplicates_columns]))
print(f'{br}ROC AUC Score: {roc_score_noduplicates}')
```

### Таким образом, простыми гипотезами отбросили довольно много признаков.

```python
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier(penalty='elasticnet', l1_ratio=1, max_iter=1000, n_jobs=-1)

sgd.fit(X_train, y_train)

roc_sgd = roc_auc_score(y_test, sgd.predict(X_test))

print(f'ROC AUC Score: {roc_sgd}')
# довольно печальная оценка на SGD.
```

### Cross validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer
from sklearn.utils import shuffle

X_s, y_s = shuffle(X, y, random_state=42)
cv_roc = make_scorer(roc_auc_score)
cross_val_score(Lasso(alpha=.5), X_s.values, y_s.values, cv=3, scoring=cv_roc, n_jobs=-1).mean()
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'alpha': np.arange(.1, 10, .5)
}

gcv_lasso = GridSearchCV(Lasso(), params, scoring=cv_roc, cv=3)
gcv_lasso.fit(X_s, y_s)

gcv_lasso.best_score_, gcv_lasso.best_params_
```

```python
gcv_ = GridSearchCV(Lasso(), {'alpha': np.arange(.1, .5, .05)}, scoring=cv_roc, cv=3)
gcv_.fit(X_s, y_s)

gcv_.best_score_, gcv_.best_params_
```

```python
gcv_ = GridSearchCV(Lasso(), {'alpha': np.arange(.1, .5, .05)}, scoring=cv_roc, cv=3)
gcv_.fit(X_s[without_duplicates_columns], y_s)

gcv_.best_score_, gcv_.best_params_
```

```python
from hyperopt import hp
from hyperopt import fmin, tpe, hp, STATUS_OK, Trials
import xgboost as xgb

dtrain = xgb.DMatrix(X.values, y.values)

params = {'colsample_bytree': 0.7, 
          'eta': 0.525, 
          'gamma': 0.65, 
          'max_depth': 10, 
          'min_child_weight': 1.0, 
          'n_estimators': 20, 
          'objective': 'binary:logistic', 
          'eval_metric': 'auc',
          'silent': 1,
          'subsample': 0.7}

xgb.cv(params, dtrain, num_boost_round=100, early_stopping_rounds=5)
```

```python
def score(params):
    print("Training with params : ")
    print(params)
    num_round = int(params['n_estimators'])
    del params['n_estimators']
    dtrain = xgb.DMatrix(X_train, label=y_train)
    dvalid = xgb.DMatrix(X_test, label=y_test)
    model = xgb.train(params, dtrain, num_round)
    predictions = model.predict(dvalid)
    score = roc_auc_score(y_test, predictions)
    br = '-'*124
    print(f'{br}\n\tScore of AUC: {score}\n{br}')
    return {'loss': score, 'status': STATUS_OK}

def optimize(trials):
    space = {
             'n_estimators' : hp.quniform('n_estimators', 10, 1000, 10),
             'eta' : hp.quniform('eta', 0.1, 0.2, 0.025),
             'max_depth' : hp.choice('max_depth', np.arange(3, 14, dtype=int)),
             'min_child_weight' : hp.quniform('min_child_weight', 1, 6, 1),
             'subsample' : hp.quniform('subsample', 0.7, 1, 0.05),
             'gamma' : hp.quniform('gamma', 0.6, 1, 0.05),
             'colsample_bytree' : hp.quniform('colsample_bytree', 0.7, 1, 0.05),
             'eval_metric': 'auc',
             'objective': 'binary:logistic',
             'nthread': 4,
             'silent' : 1
             }

    best = fmin(score, space, algo=tpe.suggest, trials=trials, max_evals=250)

    print(best)
    return best

    
trials = Trials()
best_opts = optimize(trials)
```
